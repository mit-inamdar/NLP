```
import pandas as pd
import gzip
import nltk
import gensim
```

```
df1 = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz" , lines=True)
```

```
df1=df1.head(20)
```

```
nltk.download('stopwords')
from gensim.models import Word2Vec
from nltk.corpus import stopwords
stop=stopwords.words('english')
```

```
punct = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
def punc(str):
  no_punc = ""
  for char in str:
    if char not in punct:
      no_punc = no_punc + char
  return no_punc
```

```
df1['reviewText']=df1['reviewText'].apply(punc)
df1['reviewText']=df1['reviewText'].str.lower().str.split()
df1['reviewText']=df1['reviewText'].apply(lambda x: [item for item in x if item not in stop])
```

```
df1.head()
```

```
def unique(text):
  a=[]
  for word in text:
    if word not in a:
      a.append(word)
  return a
```

```
df1['reviewText_unique']=df1['reviewText'].apply(unique)
df1.head()
```

```
from itertools import combinations
```

```
def comb(text):
  return(list(combinations(text,2)))
```

```
df1['reviewText_comb']=df1['reviewText_unique'].apply(comb)
```

```
list = [[word] for word in df1['reviewText_unique'][0]]
list
```

```
model=gensim.models.Word2Vec(list, window=5, min_count=1, workers=4)
```

```
for i in list:
  for j in list:
    if j>i:
      print(i[0],j[0],model.similarity(i[0],j[0]))
    else:
      continue
```
